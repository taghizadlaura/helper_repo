At this point, we want to link our data with real-world events around the US presidential election of 2016. We first use Google Trends to analyse what had happened, e.g., major events, protests, movies, people, ...
### 3.4.1 Get gloabal events from Google Trends
First, we analyse the topic "United States presidential election" using Google Trends data from January 1, 2014 to April 30, 2017.

```python
from pytrends.request import TrendReq
import matplotlib.pyplot as plt
```

```python
pytrends = TrendReq(hl='en-US', tz=360)

kw_US_presidential_election = "/m/07wcy"
pytrends.build_payload([kw_US_presidential_election], timeframe="2014-01-01 2017-04-30")
df = pytrends.interest_over_time()

plt.figure(figsize=(14,6))
plt.plot(df.index, df[kw_US_presidential_election])
plt.title("Google Trends - Topic: United States presidential election")
plt.grid(True)
plt.show()
```

We find peaks in 2016 with the biggest one around November 6 - 12, 2016. From the related topics, we found the following people, movies and their plots.
